```python
# 1% Open
import numpy as np
import matplotlib.pyplot as plt

#    
files = [
    
    r"C:\Users\User\Desktop\Solar Cell Simulation by PySpice\Data for Paper\Shunt & Open\0.25.txt",
    r"C:\Users\User\Desktop\Solar Cell Simulation by PySpice\Data for Paper\Shunt & Open\0.5.txt",
    r"C:\Users\User\Desktop\Solar Cell Simulation by PySpice\Data for Paper\Shunt & Open\0.75.txt",
    r"C:\Users\User\Desktop\Solar Cell Simulation by PySpice\Data for Paper\Shunt & Open\1.txt",
    r"C:\Users\User\Desktop\Solar Cell Simulation by PySpice\Data for Paper\Shunt & Open\1.25.txt",
    r"C:\Users\User\Desktop\Solar Cell Simulation by PySpice\Data for Paper\Shunt & Open\1.5.txt",
    r"C:\Users\User\Desktop\Solar Cell Simulation by PySpice\Data for Paper\Shunt & Open\1.75.txt",
    r"C:\Users\User\Desktop\Solar Cell Simulation by PySpice\Data for Paper\Shunt & Open\2.txt",
    
    # ...    
]


#     
for file_path in files:
    with open(file_path, 'r') as file:
        data = [float(line.strip()) for line in file]
    
    #   (:     1ms )
    time = np.linspace(0, len(data) * 0.001, len(data))
    
    #   
    plt.plot(time, data, label=file_path)  # label     

#  
plt.xlabel('Voltage(V)')
plt.ylabel('Current Density(A/cm)')
plt.grid(True)
plt.ylim(0, 0.02)
plt.show()
```

```python
# 1% Open
import matplotlib.pyplot as plt
plt.rcParams['axes.titlesize'] = 20  # 
plt.rcParams['axes.labelsize'] = 15  #  
plt.rcParams['xtick.labelsize'] = 13  # x 
plt.rcParams['ytick.labelsize'] = 13  # y 
plt.rcParams['legend.fontsize'] = 13  # 

x_data =[0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2]


y_data1 = [1.915, 1.912, 1.909, 1.906, 1.904, 1.901, 1.898, 1.895]
y_data2 = [19.22, 19.24, 19.25, 19.26, 19.28, 19.29, 19.29, 19.29]
y_data3 = [83.83, 78.34, 75.61, 73.1, 70.39, 67.76, 65.13, 62.52]
y_data4 = [29.75, 28.82, 27.78, 26.84, 25.85, 24.85, 23.85, 22.86]

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.plot(x_data, y_data1, 'o-r')  # 
plt.xlabel('Opened fraction(%)')
plt.ylabel('V_oc(V)')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(x_data, y_data2, 'o-b')  # 
plt.xlabel('Opened fraction(%)')
plt.ylabel('Current Density(mA/cm)')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(x_data, y_data3, 'o-g')  # 
plt.xlabel('Opened fraction(%)')
plt.ylabel('Fill Factor(%)')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(x_data, y_data4, 'o-c')  # 
plt.xlabel('Opened fraction(%)')
plt.ylabel('Power Conversion Efficiency(%)')
plt.grid(True)

plt.tight_layout()  
plt.show()
```
