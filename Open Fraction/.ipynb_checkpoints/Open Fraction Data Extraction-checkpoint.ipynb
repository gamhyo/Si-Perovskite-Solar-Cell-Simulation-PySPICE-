```python
import matplotlib.pyplot as plt
plt.rcParams['axes.titlesize'] = 20  # 제목
plt.rcParams['axes.labelsize'] = 15  # 축 이름
plt.rcParams['xtick.labelsize'] = 13  # x축 눈금
plt.rcParams['ytick.labelsize'] = 13  # y축 눈금
plt.rcParams['legend.fontsize'] = 13  # 범례

x_data =[0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2]


y_data1 = [1.917, 1.917, 1.917, 1.917, 1.917, 1.917, 1.917, 1.917, 1.917]
y_data2 = [19.32, 19.3, 19.26, 19.23, 19.19, 19.15, 19.11, 19.07, 19.03]
y_data3 = [83.47, 83.45, 83.44, 83.45, 83.46, 83.47, 83.5, 83.53, 83.55]
y_data4 = [30.92, 30.87, 30.81, 30.76, 30.71, 30.65, 30.59, 30.53, 30.48]

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.plot(x_data, y_data1, 'o-r')  # 빨간색
plt.xlabel('Opened fraction(%)')
plt.ylabel('V_oc(V)')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(x_data, y_data2, 'o-b')  # 파란색
plt.xlabel('Opened fraction(%)')
plt.ylabel('Current Density(mA/cm²)')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(x_data, y_data3, 'o-g')  # 초록색
plt.xlabel('Opened fraction(%)')
plt.ylabel('Fill Factor(%)')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(x_data, y_data4, 'o-c')  # 청록색
plt.xlabel('Opened fraction(%)')
plt.ylabel('Power Conversion Efficiency(%)')
plt.grid(True)

plt.tight_layout()  
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams['axes.titlesize'] = 20  
plt.rcParams['axes.labelsize'] = 20  
plt.rcParams['xtick.labelsize'] = 18  
plt.rcParams['ytick.labelsize'] = 18  
plt.rcParams['legend.fontsize'] = 18  

x_data =[0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2]

y_data1 = [1.917, 1.917, 1.917, 1.917, 1.917, 1.917, 1.917, 1.917, 1.917]
y_data2 = [19.32, 19.3, 19.26, 19.23, 19.19, 19.15, 19.11, 19.07, 19.03]
y_data3 = [83.47, 83.45, 83.44, 83.45, 83.46, 83.47, 83.5, 83.53, 83.55]
y_data4 = [30.92, 30.87, 30.81, 30.76, 30.71, 30.65, 30.59, 30.53, 30.48]

fig, ax1 = plt.subplots(figsize=(15, 10))

color = 'tab:red'
ax1.set_xlabel('Area Fraction of Open Defects(%)', fontsize=20)
ax1.set_ylabel('V_oc(V)', color=color, fontsize=18, fontweight='bold')
ax1.plot(x_data, y_data1, 'o-r')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Current Density(mA/cm²)', color=color, fontsize=18, fontweight='bold')
ax2.plot(x_data, y_data2, 'o-b')
ax2.tick_params(axis='y', labelcolor=color)

ax3 = ax1.twinx()
color = 'tab:green'
ax3.spines['right'].set_position(('outward', 80))  
ax3.set_ylabel('Fill Factor(%)', color=color, fontsize=18, fontweight='bold')
ax3.plot(x_data, y_data3, 'o-g')
ax3.tick_params(axis='y', labelcolor=color)

ax4 = ax1.twinx()
color = 'tab:cyan'
ax4.spines['right'].set_position(('outward', 160))  
ax4.set_ylabel('Power Conversion Efficiency(%)', color=color, fontsize=18, fontweight='bold')
ax4.plot(x_data, y_data4, 'o-c')
ax4.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.show()
```
