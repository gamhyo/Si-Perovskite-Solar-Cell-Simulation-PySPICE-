```python
import numpy as np
import matplotlib.pyplot as plt

#    
files = [
    r"C:\Users\User\Desktop\Solar Cell Simulation by PySpice\Data for Paper\Shunt Fraction\0.txt",
    r"C:\Users\User\Desktop\Solar Cell Simulation by PySpice\Data for Paper\Shunt Fraction\0.25.txt",
    r"C:\Users\User\Desktop\Solar Cell Simulation by PySpice\Data for Paper\Shunt Fraction\0.5.txt",
    r"C:\Users\User\Desktop\Solar Cell Simulation by PySpice\Data for Paper\Shunt Fraction\0.75.txt",
    r"C:\Users\User\Desktop\Solar Cell Simulation by PySpice\Data for Paper\Shunt Fraction\1.txt",
    r"C:\Users\User\Desktop\Solar Cell Simulation by PySpice\Data for Paper\Shunt Fraction\1.25.txt",
    r"C:\Users\User\Desktop\Solar Cell Simulation by PySpice\Data for Paper\Shunt Fraction\1.5.txt",
    r"C:\Users\User\Desktop\Solar Cell Simulation by PySpice\Data for Paper\Shunt Fraction\1.75.txt",
    r"C:\Users\User\Desktop\Solar Cell Simulation by PySpice\Data for Paper\Shunt Fraction\2.txt",
    
    # ...    
]


#     
for file_path in files:
    with open(file_path, 'r') as file:
        data = [float(line.strip()) for line in file]
    
    #   (:     1ms )
    time = np.linspace(0, len(data) * 0.001, len(data))
    
    #   
    plt.plot(time, data, label=file_path)  # label     

#  
plt.xlabel('Voltage(V)')
plt.ylabel('Current Density(A/cm)')
plt.grid(True)
plt.ylim(0, 0.02)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['axes.titlesize'] = 20  # 
plt.rcParams['axes.labelsize'] = 15  #  
plt.rcParams['xtick.labelsize'] = 13  # x 
plt.rcParams['ytick.labelsize'] = 13  # y 
plt.rcParams['legend.fontsize'] = 13  # 

x_data =[0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2]


y_data1 = [1.917, 1.915, 1.912, 1.909, 1.906, 1.904, 1.901, 1.898, 1.895]
y_data2 = [19.32, 19.32, 19.32, 19.32, 19.32, 19.32, 19.32, 19.32, 19.32]
y_data3 = [83.47, 80.99, 78.5, 75.98, 73.45, 70.78, 68.17, 65.58, 62.96]
y_data4 = [30.92, 29.97, 29, 28.03, 27.05, 26.04, 25.04, 24.05, 23.05]

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.plot(x_data, y_data1, 'o-r')  # 
plt.xlabel('Opened fraction(%)')
plt.ylabel('V_oc(V)')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(x_data, y_data2, 'o-b')  # 
plt.xlabel('Opened fraction(%)')
plt.ylabel('Current Density(mA/cm)')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(x_data, y_data3, 'o-g')  # 
plt.xlabel('Opened fraction(%)')
plt.ylabel('Fill Factor(%)')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(x_data, y_data4, 'o-c')  # 
plt.xlabel('Opened fraction(%)')
plt.ylabel('Power Conversion Efficiency(%)')
plt.grid(True)

plt.tight_layout()  
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams['axes.titlesize'] = 20  
plt.rcParams['axes.labelsize'] = 15  
plt.rcParams['xtick.labelsize'] = 13  
plt.rcParams['ytick.labelsize'] = 13  
plt.rcParams['legend.fontsize'] = 13  

x_data =[0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2]

y_data1 = [1.917, 1.915, 1.912, 1.909, 1.906, 1.904, 1.901, 1.898, 1.895]
y_data2 = [19.32, 19.32, 19.32, 19.32, 19.32, 19.32, 19.32, 19.32, 19.32]
y_data3 = [83.47, 80.99, 78.5, 75.98, 73.45, 70.78, 68.17, 65.58, 62.96]
y_data4 = [30.92, 29.97, 29, 28.03, 27.05, 26.04, 25.04, 24.05, 23.05]
fig, ax1 = plt.subplots(figsize=(15, 10))

color = 'tab:red'
ax1.set_xlabel('Shunted fraction(%)')
ax1.set_ylabel('V_oc(V)', color=color)
ax1.plot(x_data, y_data1, 'o-r')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Current Density(mA/cm)', color=color)
ax2.plot(x_data, y_data2, 'o-b')
ax2.tick_params(axis='y', labelcolor=color)

ax3 = ax1.twinx()
color = 'tab:green'
ax3.spines['right'].set_position(('outward', 80))  
ax3.set_ylabel('Fill Factor(%)', color=color)
ax3.plot(x_data, y_data3, 'o-g')
ax3.tick_params(axis='y', labelcolor=color)

ax4 = ax1.twinx()
color = 'tab:cyan'
ax4.spines['right'].set_position(('outward', 160))  
ax4.set_ylabel('Power Conversion Efficiency(%)', color=color)
ax4.plot(x_data, y_data4, 'o-c')
ax4.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.show()
```
