```python
import numpy as np
import matplotlib.pyplot as plt

# 불러올 파일의 경로 리스트
files = [
    r"C:\Users\jfkn5\OneDrive - 한양대학교\바탕 화면\Jeon Gamhyo\Solar Cell Simulation by PySpice\Data for Paper\Shunt Fraction\0.txt",
    r"C:\Users\jfkn5\OneDrive - 한양대학교\바탕 화면\Jeon Gamhyo\Solar Cell Simulation by PySpice\Data for Paper\Shunt Fraction\0.25.txt",
    r"C:\Users\jfkn5\OneDrive - 한양대학교\바탕 화면\Jeon Gamhyo\Solar Cell Simulation by PySpice\Data for Paper\Shunt Fraction\0.5.txt",
    r"C:\Users\jfkn5\OneDrive - 한양대학교\바탕 화면\Jeon Gamhyo\Solar Cell Simulation by PySpice\Data for Paper\Shunt Fraction\0.75.txt",
    r"C:\Users\jfkn5\OneDrive - 한양대학교\바탕 화면\Jeon Gamhyo\Solar Cell Simulation by PySpice\Data for Paper\Shunt Fraction\1.txt",
    r"C:\Users\jfkn5\OneDrive - 한양대학교\바탕 화면\Jeon Gamhyo\Solar Cell Simulation by PySpice\Data for Paper\Shunt Fraction\1.25.txt",
    r"C:\Users\jfkn5\OneDrive - 한양대학교\바탕 화면\Jeon Gamhyo\Solar Cell Simulation by PySpice\Data for Paper\Shunt Fraction\1.5.txt",
    r"C:\Users\jfkn5\OneDrive - 한양대학교\바탕 화면\Jeon Gamhyo\Solar Cell Simulation by PySpice\Data for Paper\Shunt Fraction\1.75.txt",
    r"C:\Users\jfkn5\OneDrive - 한양대학교\바탕 화면\Jeon Gamhyo\Solar Cell Simulation by PySpice\Data for Paper\Shunt Fraction\2.txt",
    
    # ... 다른 파일 경로 추가
]

#각 파일을 반복문을 통해 읽어들임
for file_path in files:
    with open(file_path, 'r') as file:
        # 데이터를 1000으로 곱해 mV 단위로 전환
        data = [float(line.strip()) * 1000 for line in file]  # 데이터 값 변환
    
    # 시간 축 (예: 데이터 포인트 간의 간격이 1ms라고 가정)
    time = np.linspace(0, len(data) * 0.001, len(data))
    
    # 데이터를 그래프에 추가, 파일 이름만 라벨로 사용
    plt.plot(time, data, label=file_path.split('\\')[-1], linewidth = 3)  # 경로가 아닌, 파일 이름만 추출하여 라벨로 사용

# 그래프 설정
plt.xlabel('Applied Voltage (V)', fontsize = 20)
# y축 라벨 mA/cm²단위로 변경
plt.ylabel('Isc (mA/cm²)', fontsize = 20)

# 눈금 글씨 조절
plt.tick_params(axis='both', which='major', labelsize=20)

plt.grid()
# y축의 값에 맞게 변경 해야 할 수도 있음
plt.ylim(0, 20)  # 예시로 수정한 값, 실제 데이터에 맞게 조정 필요

# 그래프를 이미지 파일로 저장, dpi로 해상도 지정 가능
plt.savefig("Open fraction trends1.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams['axes.titlesize'] = 30
plt.rcParams['axes.labelsize'] = 30
plt.rcParams['xtick.labelsize'] = 25
plt.rcParams['ytick.labelsize'] = 25
plt.rcParams['legend.fontsize'] = 25

x_data =[0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2]

y_data1 = [1.917, 1.915, 1.912, 1.909, 1.906, 1.904, 1.901, 1.898, 1.895]
y_data2 = [19.32, 19.32, 19.32, 19.32, 19.32, 19.32, 19.32, 19.32, 19.32]
y_data3 = [83.47, 80.99, 78.5, 75.98, 73.45, 70.78, 68.17, 65.58, 62.96]
y_data4 = [30.92, 29.97, 29, 28.03, 27.05, 26.04, 25.04, 24.05, 23.05]
fig, ax1 = plt.subplots(figsize=(20, 15))

color = 'tab:red'
ax1.set_xlabel('Area Fraction of Shunt Defects(%)', fontweight='bold')
ax1.set_ylabel('V_oc(V)', color=color, fontweight='bold')
ax1.plot(x_data, y_data1, 'o-r', linewidth=5, markersize=20)  # 여기를 수정
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Current Density(mA/cm²)', color=color, fontweight='bold')
ax2.plot(x_data, y_data2, 'o-b', linewidth=5, markersize=20)  # 여기를 수정
ax2.tick_params(axis='y', labelcolor=color)

ax3 = ax1.twinx()
color = 'tab:green'
ax3.spines['right'].set_position(('outward', 120))
ax3.set_ylabel('Fill Factor(%)', color=color, fontweight='bold')
ax3.plot(x_data, y_data3, 'o-g', linewidth=5, markersize=20)  # 여기를 수정
ax3.tick_params(axis='y', labelcolor=color)

ax4 = ax1.twinx()
color = 'tab:cyan'
ax4.spines['right'].set_position(('outward', 240))
ax4.set_ylabel('Power Conversion Efficiency(%)', color=color, fontweight='bold')
ax4.plot(x_data, y_data4, 'o-c', linewidth=5, markersize=20)  # 여기를 수정
ax4.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.show()

```
