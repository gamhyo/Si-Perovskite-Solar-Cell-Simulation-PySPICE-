```python
import matplotlib.pyplot as plt

plt.rcParams['axes.titlesize'] = 30
plt.rcParams['axes.labelsize'] = 30
plt.rcParams['xtick.labelsize'] = 25
plt.rcParams['ytick.labelsize'] = 25
plt.rcParams['legend.fontsize'] = 25

x_data =[0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2]

y_data1 = [1.917, 1.915, 1.912, 1.909, 1.906, 1.904, 1.901, 1.898, 1.895]
y_data2 = [19.32, 19.32, 19.32, 19.32, 19.32, 19.32, 19.32, 19.32, 19.32]
y_data3 = [83.47, 80.99, 78.5, 75.98, 73.45, 70.78, 68.17, 65.58, 62.96]
y_data4 = [30.92, 29.97, 29, 28.03, 27.05, 26.04, 25.04, 24.05, 23.05]
fig, ax1 = plt.subplots(figsize=(20, 15))

color = 'tab:red'
ax1.set_xlabel('Area Fraction of Shunt Defects(%)', fontweight='bold')
ax1.set_ylabel('V_oc(V)', color=color, fontweight='bold')
ax1.plot(x_data, y_data1, 'o-r', linewidth=5, markersize=20)  #  
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Current Density(mA/cm)', color=color, fontweight='bold')
ax2.plot(x_data, y_data2, 'o-b', linewidth=5, markersize=20)  #  
ax2.tick_params(axis='y', labelcolor=color)

ax3 = ax1.twinx()
color = 'tab:green'
ax3.spines['right'].set_position(('outward', 120))
ax3.set_ylabel('Fill Factor(%)', color=color, fontweight='bold')
ax3.plot(x_data, y_data3, 'o-g', linewidth=5, markersize=20)  #  
ax3.tick_params(axis='y', labelcolor=color)

ax4 = ax1.twinx()
color = 'tab:cyan'
ax4.spines['right'].set_position(('outward', 240))
ax4.set_ylabel('Power Conversion Efficiency(%)', color=color, fontweight='bold')
ax4.plot(x_data, y_data4, 'o-c', linewidth=5, markersize=20)  #  
ax4.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.show()

```
