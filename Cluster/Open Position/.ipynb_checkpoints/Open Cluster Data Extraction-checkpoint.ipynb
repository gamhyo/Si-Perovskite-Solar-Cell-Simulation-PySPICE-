```python
import matplotlib.pyplot as plt

plt.rcParams['axes.titlesize'] = 30
plt.rcParams['axes.labelsize'] = 30
plt.rcParams['xtick.labelsize'] = 25
plt.rcParams['ytick.labelsize'] = 25
plt.rcParams['legend.fontsize'] = 25

x_data =["Near", "..","..." , "Far" ]

y_data1 = [1.917, 1.917, 1.917, 1.917]
y_data2 = [19.2, 19.2, 19.2, 19.2 ]
y_data3 = [83.5, 83.51, 83.51, 83.51]
y_data4 = [30.73, 30.73, 30.73, 30.73]

y_data11 = [1.908, 1.909, 1.909, 1.909]
y_data22 = [19.32, 19.32, 19.32, 19.32]
y_data33 = [76.79, 76.75, 76.75, 76.77]
y_data44 = [28.31, 28.31, 28.31, 28.32]

y_data5 = [1.917, 1.917, 1.917, 1.917]
y_data6 = [19.2, 19.2, 19.2, 19.2 ]
y_data7 = [83.5, 83.5, 83.5, 83.5]
y_data8 = [30.73, 30.73, 30.73, 30.73]

y_data55= [1.909, 1.909, 1.909, 1.909]
y_data66= [19.32, 19.32, 19.32, 19.32]
y_data77= [76.66, 76.66, 76.7, 76.8]
y_data88= [28.28, 28.28, 28.29, 28.33]



plt.figure(figsize=(15, 10))

from mpl_toolkits.axes_grid1.inset_locator import inset_axes

plt.figure(figsize=(15, 10))

def plot_with_broken_y_axis(ax, x_data, y_data_list, y_labels, title):
    # Main plot
    for y_data, label in zip(y_data_list, y_labels):
        ax.plot(x_data, y_data, 'o-', linewidth=5, markersize=20, label=label)
    ax.set_xlabel('Distance from Top Grid', fontsize=18, fontweight='bold')
    ax.set_title(title, fontsize=20, fontweight='bold')
    ax.legend(fontsize=18)
    ax.grid(True)

    # Inserted plot
    ax2 = inset_axes(ax, width='100%', height='30%', loc='upper center', bbox_to_anchor=(0., 0.1, 1, 1), bbox_transform=ax.transAxes)
    for y_data, label in zip(y_data_list, y_labels):
        ax2.plot(x_data, y_data, 'o-', linewidth=5, markersize=20, label=label)
    ax2.grid(True)
    
    return ax, ax2

ax1 = plt.subplot(2, 2, 1)
plot_with_broken_y_axis(ax1, x_data, [y_data1, y_data11, y_data5, y_data55], ['Left to Right(Open)', 'Left to Right(Shunt)', 'Down to Up(Open)', 'Down to Up(Shunt)'], 'V_oc(V)')

ax2 = plt.subplot(2, 2, 2)
plot_with_broken_y_axis(ax2, x_data, [y_data2, y_data22, y_data6, y_data66], ['Left to Right(Open)', 'Left to Right(Shunt)', 'Down to Up(Open)', 'Down to Up(Shunt)'], 'Current Density(mA/cm)')

ax3 = plt.subplot(2, 2, 3)
plot_with_broken_y_axis(ax3, x_data, [y_data3, y_data33, y_data7, y_data77], ['Left to Right(Open)', 'Left to Right(Shunt)', 'Down to Up(Open)', 'Down to Up(Shunt)'], 'Fill Factor(%)')

ax4 = plt.subplot(2, 2, 4)
plot_with_broken_y_axis(ax4, x_data, [y_data4, y_data44, y_data8, y_data88], ['Left to Right(Open)', 'Left to Right(Shunt)', 'Down to Up(Open)', 'Down to Up(Shunt)'], 'Power Conversion Efficiency(%)')

plt.tight_layout()
plt.show()


plt.tight_layout()  #     .
plt.show()
```
