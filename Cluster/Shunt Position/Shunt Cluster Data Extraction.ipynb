```python
import matplotlib.pyplot as plt
plt.rcParams['axes.titlesize'] = 20  # 제목
plt.rcParams['axes.labelsize'] = 15  # 축 이름
plt.rcParams['xtick.labelsize'] = 13  # x축 눈금
plt.rcParams['ytick.labelsize'] = 13  # y축 눈금
plt.rcParams['legend.fontsize'] = 13  # 범례

x_data =["Near", "..","..." , "Far" ]

y_data1 = [1.908, 1.909, 1.909, 1.909]
y_data2 = [19.32, 19.32, 19.32, 19.32]
y_data3 = [76.79, 76.75, 76.75, 76.77]
y_data4 = [28.31, 28.31, 28.31, 28.32]


y_data5= [1.909, 1.909, 1.909, 1.909]
y_data6= [19.32, 19.32, 19.32, 19.32]
y_data7= [76.66, 76.66, 76.7, 76.8]
y_data8= [28.28, 28.28, 28.29, 28.33]


plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.plot(x_data, y_data1, 'o-', linewidth=5, markersize=20, label='Left to Right')
plt.plot(x_data, y_data5, 'o-', linewidth=5, markersize=20, label='Down to Up')
plt.xlabel('Distance from Top Grid', fontsize=20, fontweight='bold')
plt.ylabel('V_oc(V)', fontsize=20, fontweight='bold')
plt.legend(fontsize=18)
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(x_data, y_data2, 'o-', linewidth=5, markersize=20, label='Left to Right')
plt.plot(x_data, y_data6, 'o-', linewidth=5, markersize=20, label='Down to Up')
plt.xlabel('Distance from Top Grid', fontsize=18, fontweight='bold')
plt.ylabel('Current Density(mA/cm²)', fontsize=18, fontweight='bold')
plt.legend(fontsize=18)
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(x_data, y_data3,'o-',  linewidth=5, markersize=20, label='Left to Right')
plt.plot(x_data, y_data7, 'o-', linewidth=5, markersize=20, label='Down to Up')
plt.xlabel('Distance from Top Grid', fontsize=20, fontweight='bold')
plt.ylabel('Fill Factor(%)', fontsize=20, fontweight='bold')
plt.legend(fontsize=18)
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(x_data, y_data4, 'o-', linewidth=5, markersize=20, label='Left to Right')
plt.plot(x_data, y_data8, 'o-', linewidth=5, markersize=20, label='Down to Up')
plt.xlabel('Distance from Top Grid', fontsize=20, fontweight='bold')
plt.ylabel('Power Conversion Efficiency(%)', fontsize=20, fontweight='bold')
plt.legend(fontsize=18)
plt.grid(True)

plt.tight_layout()  # 그래프 간의 간격을 자동으로 맞춰줍니다.
plt.show()


```
