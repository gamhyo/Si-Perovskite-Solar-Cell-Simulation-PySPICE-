```python
import matplotlib.pyplot as plt
plt.rcParams['axes.titlesize'] = 20  # 
plt.rcParams['axes.labelsize'] = 15  #  
plt.rcParams['xtick.labelsize'] = 13  # x 
plt.rcParams['ytick.labelsize'] = 13  # y 
plt.rcParams['legend.fontsize'] = 13  # 

x_data =["Near", ".",".." , "Far" ]

y_data1 = [1.908, 1.909, 1.909, 1.909]
y_data2 = [19.32, 19.32, 19.32, 19.32]
y_data3 = [76.79, 76.75, 76.75, 76.77]
y_data4 = [28.31, 28.31, 28.31, 28.32]


y_data5= [1.909, 1.909, 1.909, 1.909]
y_data6= [19.32, 19.32, 19.32, 19.32]
y_data7= [76.66, 76.66, 76.7, 76.8]
y_data8= [28.28, 28.28, 28.29, 28.33]

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.plot(x_data, y_data1, 'o-')
plt.plot(x_data, y_data5, 'o-')
plt.xlabel('Distance from Top Grid')
plt.ylabel('V_oc(V)')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(x_data, y_data2, 'o-')
plt.plot(x_data, y_data6, 'o-')
plt.xlabel('Distance from Top Grid')
plt.ylabel('Current Density(mA/cm)')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(x_data, y_data3, 'o-') 
plt.plot(x_data, y_data6, 'o-')
plt.xlabel('Distance from Top Grid')
plt.ylabel('Fill Factor(%)')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(x_data, y_data4, 'o-') 
plt.plot(x_data, y_data8, 'o-') 
plt.xlabel('Distance from Top Grid')
plt.ylabel('Power Conversion Efficiency(%)')
plt.grid(True)

plt.tight_layout()  
plt.show()


```
