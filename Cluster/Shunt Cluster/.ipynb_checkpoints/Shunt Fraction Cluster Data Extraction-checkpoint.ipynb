```python
import matplotlib.pyplot as plt
plt.rcParams['axes.titlesize'] = 20  # 제목
plt.rcParams['axes.labelsize'] = 15  # 축 이름
plt.rcParams['xtick.labelsize'] = 13  # x축 눈금
plt.rcParams['ytick.labelsize'] = 13  # y축 눈금
plt.rcParams['legend.fontsize'] = 13  # 범례

x_data =[144, 288, 432, 576]


y_data1 = [1.913, 1.912, 1.912, 1.912]
y_data2 = [19.32, 19.32, 19.32, 19.32]
y_data3 = [79.16, 79.06, 79.03, 79.02]
y_data4 = [29.26, 29.21, 29.2, 29.19]

y_data5 = [1.908, 1.908, 1.907, 1.907]
y_data6 = [19.32, 19.32, 19.32, 19.32]
y_data7 = [74.75, 74.52, 74.55, 74.54]
y_data8 = [27.56, 27.48, 27.47, 27.47]

y_data9 = [1.903, 1.903, 1.902, 1.902]
y_data10 = [19.32, 19.32, 19.32, 19.32]
y_data11 = [70.24, 70.01, 70.08, 69.69]
y_data12 = [25.83, 25.74, 25.76, 25.61]

y_data13 = [1.898, 1.897, 1.897, 1.897]
y_data14 = [19.32, 19.32, 19.32, 19.32]
y_data15 = [65.72, 65.57, 65.6, 65.65]
y_data16 = [24.1, 24.04, 24.05, 24.07]


plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.plot(x_data, y_data1, 'o-')
plt.plot(x_data, y_data5, 'o-')
plt.plot(x_data, y_data9, 'o-')
plt.plot(x_data, y_data13, 'o-')
plt.xlabel('# of shunted pixels')
plt.ylabel('V_oc(V)')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(x_data, y_data2, 'o-')
plt.plot(x_data, y_data6, 'o-')
plt.plot(x_data, y_data10, 'o-')
plt.plot(x_data, y_data14, 'o-')
plt.xlabel('# of shunted pixels')
plt.ylabel('Current Density(mA/cm²)')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(x_data, y_data3,'o-')
plt.plot(x_data, y_data7, 'o-')
plt.plot(x_data, y_data11, 'o-')
plt.plot(x_data, y_data15, 'o-')
plt.xlabel('# of shunted pixels')
plt.ylabel('Fill Factor(%)')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(x_data, y_data4, 'o-')
plt.plot(x_data, y_data8, 'o-')
plt.plot(x_data, y_data12, 'o-')
plt.plot(x_data, y_data16, 'o-')
plt.xlabel('# of shunted pixels')
plt.ylabel('Power Conversion Efficiency(%)')
plt.grid(True)

plt.tight_layout()  # 그래프 간의 간격을 자동으로 맞춰줍니다.
plt.show()
```
