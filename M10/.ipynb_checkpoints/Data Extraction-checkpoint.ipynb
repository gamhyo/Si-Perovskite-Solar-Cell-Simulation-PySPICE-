```python
import matplotlib.pyplot as plt

plt.rcParams['axes.titlesize'] = 20  
plt.rcParams['axes.labelsize'] = 15  
plt.rcParams['xtick.labelsize'] = 13  
plt.rcParams['ytick.labelsize'] = 13  
plt.rcParams['legend.fontsize'] = 13  

x_data = [0, 1, 2]

#   
y_data_sets = [
    ("Data1", [30.8595, 30.6615, 30.4193]),
    ("Data2", [0.5323, 0.5323, 0.5323]),
    ("Data3", [0.9898, 0.9761, 0.9583]),
    ("Data4", [1.3023, 1.3011, 1.3109]),
    ("Data5", [0.0736, 0.072, 0.0702]),
    ("Data6", [0.0395, 0.0386, 0.0376]),
    ("Data7", [0.1278, 0.1251, 0.1219]),
    ("Data8", [0.6451, 0.6378, 0.6280]),
    ("Data9", [0.0679, 0.0669, 0.0657]),
    ("Data10", [0.036, 0.0355, 0.0348]),
    ("Data11", [0.2651, 0.2698, 0.2382]),
    ("Data12", [0.4985, 0.4931, 0.5305]),
    ("Data13", [0.5388, 0.5382, 0.5423]),
    ("Data14", [0, 0, 0])
]

# 5x3   (5 3)
fig, axs = plt.subplots(5, 3, figsize=(18, 18))

for i, (label, y_data) in enumerate(y_data_sets):
    row = i // 3
    col = i % 3
    ax = axs[row, col]
    ax.plot(x_data, y_data, label=label)
    ax.set_title(f"Trend of {label}")
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.legend()

#     
for i in range(len(y_data_sets), 5 * 3):
    row = i // 3
    col = i % 3
    axs[row, col].axis('off')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams['axes.titlesize'] = 20  
plt.rcParams['axes.labelsize'] = 15  
plt.rcParams['xtick.labelsize'] = 13  
plt.rcParams['ytick.labelsize'] = 13  
plt.rcParams['legend.fontsize'] = 10  

x_data = [0, 1, 2]

y_data_sets = [
    ("Data1", [30.8595, 30.6615, 30.4193]),
    ("Data2", [0.5323, 0.5323, 0.5323]),
    ("Data3", [0.9898, 0.9761, 0.9583]),
    ("Data4", [1.3023, 1.3011, 1.3109]),
    ("Data5", [0.0736, 0.072, 0.0702]),
    ("Data6", [0.0395, 0.0386, 0.0376]),
    ("Data7", [0.1278, 0.1251, 0.1219]),
    ("Data8", [0.6451, 0.6378, 0.6280]),
    ("Data9", [0.0679, 0.0669, 0.0657]),
    ("Data10", [0.036, 0.0355, 0.0348]),
    ("Data11", [0.2651, 0.2698, 0.2382]),
    ("Data12", [0.4985, 0.4931, 0.5305]),
    ("Data13", [0.5388, 0.5382, 0.5423]),
    ("Data14", [0, 0, 0])
]

fig, ax = plt.subplots(figsize=(20, 10))

for label, y_data in y_data_sets:
    ax.plot(x_data, y_data, label=label)

ax.set_title('Trends of Multiple Data Sets')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()

plt.show()
```
